```python
data = data.dropna(how='all')
```

```python
views_dict = dict(data.count()[1:])
```

```python
for movie,views in views_dict.items():
    if views == max(views_dict.values()):
        print(f'{movie} has maximum views/ratings')
```

```python
total_rating = []
for col in data.columns:
    if col != 'user_id':
        total_rating.append(data[col].sum())
```

```python
view_count = list(data.count()[1:])
```

```python
task_2 = list(np.array(total_rating)/np.array(view_count))
```

```python
avg_rating = {}
for i in range(len(list(data.count()[1:].index))):
    avg_rating[list(data.count()[1:].index)[i]] = task_2[i]
```

```python
five_rating = {}
for i,j in avg_rating.items():
    if j == 5.0:
         five_rating[i] = avg_rating[i]
        
```

```python
for i,j in five_rating.items():
    five_rating[i] = views_dict[i]
```

```python
top_five_rating = {k: v for k, v in sorted(five_rating.items(), key=lambda item: item[1],reverse=True)}
```

```python
top_5_movies = []
for i in range(5):
    top_5_movies.append(list(top_five_rating.keys())[i])
```
